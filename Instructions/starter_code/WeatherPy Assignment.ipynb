```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
cities_info_df = pd.DataFrame({"City":cities})
#cities_info_df.head()
```

```python
cities_info_df["Longitude"]=""
cities_info_df["Latitude"]=""
cities_info_df["Cloudiness"]=""
cities_info_df["Country"]=""
cities_info_df["Date"]=""
cities_info_df["Humidity"]=""
cities_info_df["Max Temp"]=""
cities_info_df["Country"]=""
cities_info_df["Wind Speed"]=""
row_count=0

unit = "imperical"

for index, row in cities_info_df.iterrows():

    time.sleep(1.1)
    
    url = "http://api.openweathermap.org/data/2.5/weather?q="
    
    query_url = url + str(row["City"]) + "&appid=" + api_key + "&unit=" + unit
    
    print(f"Now retrieving city #{row_count}")
    print("----------------------------------")
    row_count += 1
    
    city_weather_info = requests.get(query_url).json()
    
    try:
        longitude = city_weather_info["coord"]["lon"]
        latitude = city_weather_info["coord"]["lat"]
        max_temp = city_weather_info["main"]["temp_max"]
        country = city_weather_info["sys"]["country"]
        humidity = city_weather_info["main"]["humidity"]
        cloudiness = city_weather_info["clouds"]["all"]
        wind_speed = city_weather_info["wind"]["speed"]
        
        cities_info_df.set_value(index, "Longitude", longitude )
        cities_info_df.set_value(index, "Latitude", latitude)
        cities_info_df.set_value(index, "Max Temp", max_temp)
        cities_info_df.set_value(index, "Country", country)
        cities_info_df.set_value(index, "Humidity", humidity)
        cities_info_df.set_value(index, "Cloudiness", cloudiness)
        cities_info_df.set_value(index, "Wind Speed", wind_speed)
        
    except:
        print("Error with Weather Data, skipping.")

            
```
